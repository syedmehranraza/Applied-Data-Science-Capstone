# FEATURE **EXTRACTION**

```python
def feature_extraction(transformed_data):
    Data = []
    for img1 in transformed_data:        
        # Applying Principal Component Analysis
        pca = PCA(n_components = 10)   #applying PCA
        img1 = pca.fit_transform(img1)
        
        # Applying Histogram of gradients
        fd , img1 = hog(img1, orientations = 9, pixels_per_cell = (8,8),
                             cells_per_block = (2,2),visualize = True,multichannel= False)
        
        # Applying edge detection
        img1 = cv.Canny(img1.astype(np.uint8),50,200)
        
        # Applying Laplacian of gaussian
        img1 = cv.Laplacian(img1,cv.CV_64F, ksize = 3)
        
        Data.append(img1)
    
    return np.array(Data)
```
